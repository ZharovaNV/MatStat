```python
def evaluate_preds(train_true_values, train_pred_values, val_true_values, val_pred_values):
    
    print("Train R2:\t" + str(round(r2(train_true_values, train_pred_values), 3)))
    print("Valid R2:\t" + str(round(r2(val_true_values, val_pred_values), 3)))
    
    plt.figure(figsize=(18,10))
    
    plt.subplot(121)
    sns.scatterplot(x=train_pred_values, y=train_true_values)
    plt.xlabel('Predicted values')
    plt.ylabel('True values')
    plt.title('Train sample prediction')
    
    plt.subplot(122)
    sns.scatterplot(x=val_pred_values, y=val_true_values)
    plt.xlabel('Predicted values')
    plt.ylabel('True values')
    plt.title('Test sample prediction')
    plt.show()
```

### Расчет матричным методом

```python
n = X_train.shape[0]
m = X_train.shape[1]
n,m
```

```python
X_ = np.hstack([np.ones((n,1)), X_train])
y_ = y_train
B_ = np.dot(np.linalg.inv(np.dot(X_.T, X_)), X_.T@y_)
```

Проверим, что получили те же значения коэффициентов:

```python
pd.DataFrame(np.insert(lr.coef_, 0, lr.intercept_), B_)
```

```python
y_train_pred_ = B_[0]+ B_[1]*X_train.iloc[:,0]+ B_[2]*X_train.iloc[:,1]+ B_[3]*X_train.iloc[:,2]
y_valid_pred_ = B_[0]+ B_[1]*X_valid.iloc[:,0]+ B_[2]*X_valid.iloc[:,1]+ B_[3]*X_valid.iloc[:,2]

evaluate_preds(y_train, y_train_pred_, y_valid, y_valid_pred_)
```

### Применение критерия Фишера

Проверим значимость модели в целом. Для этого сформулируем гипотезу $H_0:$ все параметры равны нулю.

Рассчитаем значение статистики Фишера и сравним его с критическим значением, взятым из таблиц распределения Фишера.

Для модели до стандартизации:

```python
SSf = ((y_train_pred_scaled_ - y_train.mean())**2).sum()
SSo = ((y_train - y_train_pred_scaled_)**2).sum()
Fp = (SSf / 3) / (SSo / (len(y_train) - 4))
Fp
```

Ft = 2.62 на уровне значимости 0.05

Fp >> Ft - модель значима на уровне значимости 0.05

Для стандартизованной модели:

```python
SSf = ((y_valid_pred_scaled_ - y_valid.mean())**2).sum()
SSo = ((y_valid - y_valid_pred_scaled_)**2).sum()
Fp = (SSf / 3) / (SSo / (len(y_valid) - 4))
Fp
```

Fp >> Ft - модель значима на уровне значимости 0.05

### Оценка значимости параметров модели

Несмещенную оценку остаточной дисперсии найдем по формуле $\hat {S}^2 = \frac 1{n-m-1} \sum (y_i - \hat(y)_i)^2$

```python
S2 = 1/(n-m-1)*((y_train -  y_train_pred_)**2).sum()
S2
```

Значимость коэффициентов регрессии проверяется с помощью t-критерия Стьюдента. 

Для этого рассчитаем статистику $t_{vbl_j} = \frac {b_j}{\hat{S}_{b_j}}$, где $\hat{S}^2_{b_j} = \hat{S}^2 [(X^TX)^{-1}]_{jj}$

Наблюдаемые значения:

```python
t_vbl = np.diagonal(np.outer((S2**0.5 * np.diagonal(np.linalg.inv(np.dot(X_scaled_.T, X_scaled_)))**0.5)**(-1), B_scaled_))
t_vbl
```
